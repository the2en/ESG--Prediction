````python
def regression_acc(x, y, net):

  sum = 0
  correct = 0

  x = np.array(x, dtype=float)
  y = np.array(y, dtype=float)

  y = y.reshape(-1, 1)

  for index, _y in enumerate(y):
    pred = net.predict([x[index]])
    sum = sum + 1
    if _y == round(pred[0], 1):
      correct = correct + 1

  return correct / sum



def classifier(sample_x, sample_y, x, y, net, name):

  print(name, '\n')

  sample_x = np.array(sample_x, dtype=float)
  sample_y = np.array(sample_y, dtype=float)

  print('Sample predicted: ', np.array(net.predict([sample_x])[0]))
  print('Sample Label: ', sample_y)

  acc = net.score(x, y)
  print('accuracy: ', acc)

  print()



def predictor(sample_x, sample_y, x, y, net, name):

  print(name, '\n')

  pred = net.predict([sample_x])

  print('Sample Label: ', np.array(sample_y))
  print('Sample Predicted: ', pred[0])

  acc = regression_acc(x, y, net)

  print('accuracy: ', acc)

  print()
````
